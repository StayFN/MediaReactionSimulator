```python
import pandas as pd

df = pd.read_csv("csv\output\Report_Shorts_with_summary_stefan_sexy_experimentResults.csv")
df.head()
```

```python
import re
def extract_sex_stim(string):
    match = re.search(r'\d+', string)
    if match:
        return int(match.group())
    else:
        return None
```

```python
columns_to_update = ['low ssc characteristics', 'high ssc characteristics', 'low ssc defined', 'high ssc defined']

for column in columns_to_update:
    df[column] = df[column].apply(extract_sex_stim)

df.head()
```

```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame 'df' with the relevant columns
columns_to_plot = ['low ssc characteristics', 'high ssc characteristics', 'low ssc defined', 'high ssc defined']

# Prepare data for the box plot
data_to_plot = [df[col].dropna() for col in columns_to_plot]

# Creating the box plot with increased figure size
fig, ax = plt.subplots(figsize=(12, 8))  # Set the size here (width, height)

# Adding patch_artist=True option to fill with color
bp = ax.boxplot(data_to_plot, patch_artist=True, notch=True, medianprops={'linewidth': 2})

# Define your colors for the boxes
#colors = ['lightblue', 'lightgreen', 'lightblue', 'lightgreen']
# Colors and labels for the legend
colors = ['lightseagreen', 'seagreen', 'palegreen', 'mediumseagreen']
labels = ['Low SSC Char.', 'High SSC Char.', 'Low SSC Def.', 'High SSC Def.']

for patch, color, label in zip(bp['boxes'], colors, labels):
    patch.set_facecolor(color)
    patch.set_label(label)

# Adding legend
ax.legend(loc='upper right')

# Adding labels
ax.set_xticklabels(['Sex_Stim \n Low SSC Char.', 'Sex_Stim \n High SSC Char.', 'Sex_Stim \n Low SSC Def.', 'Sex_Stim \n High SSC Def.'])
ax.set_ylabel('Sexual Stimulation Scores')
ax.set_title('Sexual Stimulation Distribution by Category')

plt.show()
```

Thus, if we have this weird "flipped" appearance in the notched box plots, it simply means that the 1st quartile has a lower value than the confidence of the mean and vice versa for the 3rd quartile. Although it looks ugly, it's actually useful information about the (un)confidence of the median.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Define the new column names
new_column_names = {
    'low ssc characteristics_enjoyment': 'Low SSC Char. Enjoym.',
    'low ssc characteristics_appropriate': 'Low SSC Char. Appr.',
    'high ssc characteristics_enjoyment': 'High SSC Char. Enjoym.',
    'high ssc characteristics_appropriate': 'High SSC Char. Appr.',
    'low ssc defined_enjoyment': 'Low SSC Def. Enjoym.',
    'low ssc defined_appropriate': 'Low SSC Def. Appr.',
    'high ssc defined_enjoyment': 'High SSC Def. Enjoym.',
    'high ssc defined_appropriate': 'High SSC Def. Appr.'
}

# Rename the columns
df2.rename(columns=new_column_names, inplace=True)

# Update the columns to plot
columns_to_plot = list(new_column_names.values())

# Convert the columns to numeric type and prepare data for the box plot
data_to_plot = [pd.to_numeric(df2[col].dropna()) for col in columns_to_plot]

# Creating the box plot with increased figure size
fig, ax = plt.subplots(figsize=(12, 8))  # Set the size here (width, height)

# Adding patch_artist=True option to fill with color
bp = ax.boxplot(data_to_plot, patch_artist=True, notch=True, medianprops={'linewidth': 2})

# Define your colors for the boxes
colors = ['lightseagreen', 'seagreen', 'palegreen', 'mediumseagreen'] * 2  # Repeat the colors for the '_appropriate' columns

# Define labels for the boxes in the order of the DataFrame columns
labels = columns_to_plot

# Adding labels
for patch, color, label in zip(bp['boxes'], colors, labels):
    patch.set_facecolor(color)
    patch.set_label(label)

# Adding labels
ax.set_xticklabels(labels, rotation=45)  # Rotate the labels 45 degrees for better visibility
ax.set_ylabel('Values')
ax.set_title('Enjoyment / Appropriate Scores by Category')

# Add the average to the boxplot
for i, line in enumerate(bp['medians']):
    # get position data for median line
    x, _ = line.get_xydata()[1]  # top of median line
    # overlay median value
    ax.text(x, 0.02, f'Mean: {np.mean(data_to_plot[i]):.2f}', horizontalalignment='center', verticalalignment='bottom', fontsize=8, color='red')  # draw above, centered

plt.show()
```
